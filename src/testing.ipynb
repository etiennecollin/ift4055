```python
%reload_ext autoreload
%autoreload 2

from utils import *
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "figure.dpi": 400,
    "savefig.transparent": True,
    "figure.figsize": (8, 4),
})
plt.style.use("dark_background")
rng = np.random.default_rng(42)
```

```python
def plot(sampling, path):
    # Compute the PDF values
    x_values = np.linspace(-15, 15, 1000)
    function_values = function(x_values)
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, function_values, label="Gaussian Mixture PDF", color="#5969a6")
    plt.scatter(sampling, function(sampling), label="Samples", color="white")
    plt.title("Gaussian Mixture Model PDF")
    plt.xlabel("x")
    plt.ylabel("PDF")
    plt.legend()
    plt.grid(False)
    plt.savefig(path)
    plt.show()
```

```python
uniform_sampling = np.linspace(-15, 15, 25)
random_sampling = rng.uniform(-15, 15, 25)

a = rng.uniform(-15, -10, 3)
b = rng.uniform(-10, -5, 9)
c = rng.uniform(-5, 5, 2)
d = rng.uniform(5, 10, 8)
e = rng.uniform(10, 15, 3)
optimal_sampling = np.concatenate((a, b, c, d, e), axis=0)

plot(uniform_sampling, "./uniform_sampling.png")
plot(random_sampling, "./random_sampling.png")
plot(optimal_sampling, "./optimal_sampling.png")
```
